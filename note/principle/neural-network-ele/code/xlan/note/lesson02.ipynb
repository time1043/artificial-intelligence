```python
import sys
sys.path.append("..")

import numpy as np
import matplotlib.pyplot as plt
from src.dataset.v1 import get_beans
from src.viz.v1 import plot_size_toxicity
```

```python
sample_size = 100
xs, ys = get_beans(sample_size)

w =0.1
y_pre = w * xs

plot_size_toxicity(xs, ys, y_pre)
```

```python
# w = 0.1
error = (y_pre - ys) ** 2
e = np.sum(error) / sample_size
print(e)
```

```python
ws = np.arange(0, 3, 0.1)


def cost_function(w):
    y_pre = w * xs
    e = np.sum((y_pre - ys) ** 2) / sample_size
    print(f"w={str(w)}, e={str(e)}")
    return e


es = [cost_function(w) for w in ws]

plt.title("cost function", fontsize=12)
plt.xlabel("w")
plt.ylabel("e")
plt.plot(ws, es)
plt.show()
```

```python
# w = - b / 2a
w_opt = np.sum(xs * ys) / np.sum(xs * xs)
e_min = cost_function(w_opt)

y_pre = w_opt * xs

plot_size_toxicity(xs, ys, y_pre)
```
