```python
import sys
sys.path.append("..")

import numpy as np
from src.dataset.v1 import get_beans
from src.viz.v1 import plot_size_toxicity
```

```python
sample_size = 100
xs, ys = get_beans(sample_size)
print(xs)
print(ys)

plot_size_toxicity(xs, ys)
```

```python
# Beans Predict
w = 0.5  # init
y_pre = w * xs  # broadcasting
print(y_pre)

plot_size_toxicity(xs, ys, y_pre)
```

```python
# Rosenblatt
w = 0.5
alpha = 0.05
epoch = 100

# Train
for _ in range(epoch):
    for i in range(sample_size):
        x = xs[i]
        y = ys[i]
        y_pre = w * x
        error = y - y_pre
        w += x * error * alpha

# Predict
y_pre = w * xs

plot_size_toxicity(xs, ys, y_pre)
```

```python
w = 0.5
w = 0.5
alpha = 0.05
epoch = 100

# Train
for _ in range(epoch):
    y_pre = w * xs
    error = ys - y_pre
    w += alpha * np.sum(xs * error) / sample_size

# Predict
y_pre = w * xs

plot_size_toxicity(xs, ys, y_pre)
```

$$
\textbf{Stochastic Gradient Descent} \quad w = w + \alpha \cdot x \cdot (y - wx) \\
\textbf{Batch Gradient Descent} \quad w = w + \alpha \cdot \sum_{i=1}^{n} x_i(y_i - wx_i)
$$
