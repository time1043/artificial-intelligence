```python
import sys
sys.path.append("..")

import numpy as np
from src.dataset.v1 import get_beans
from src.viz.v1 import plot_size_toxicity
```

```python
sample_size = 100
xs, ys = get_beans(sample_size)

w = 0.1
y_pre = w * xs

plot_size_toxicity(xs, ys, y_pre)
```

```python
# Stochastic Gradient Descent (One sample at a time)
w = 0.1
alpha = 0.01
epoch = 100

# Train
for _ in range(epoch):
    for i in range(sample_size):
        x, y = xs[i], ys[i]
        k = 2 * x**2 * w + (-2 * x * y)
        w -= alpha * k  # w += alpha * 2x * (y - wx)

# Predict
y_pre = w * xs

plot_size_toxicity(xs, ys, y_pre)
```

```python
# Batch Gradient Descent (All samples at a time)
w = 0.1
alpha = 0.01
epoch = 300

# Train
for _ in range(epoch):
    k = np.sum(2 * xs**2 * w + -2 * xs * ys) / sample_size
    w -= alpha * k

# Predict
y_pre = w * xs

plot_size_toxicity(xs, ys, y_pre)
```
